# The Metric
The dataset I used is "<a href = "https://data.wprdc.org/dataset/city-of-pittsburgh-facilities/resource/fbb50b02-2879-47cd-abea-ae697ec05170">City of Pittsburgh Facilities</a>" from WPRDC, which is a list of all the public facilities in Pittsburgh with information about their name, location, type of facility, and more. This dataset was created in May of 2017 and was last updated 3 days ago at the time of writing this, so all of the data is very recent. First, I imported pandas and matplotlib to read the data and create graphs. I then stored the data from the website in a variable I named pf. To make sure that my code was working correctly, I displayed the first 10 sets of data in pf.

```python
import pandas as pd

import matplotlib.pyplot as plot
pf = pd.read_csv("https://data.wprdc.org/datastore/dump/fbb50b02-2879-47cd-abea-ae697ec05170") 
pf.head(10)
```

# Number of public facilities by neighborhood
There are a lot of colums in this dataset, so instead of filtering out every column that wasn't neighborhood, I created a neighborhood variable that counted the number of times each neighborhood showed up in the dataset. There are 15 neighborhoods that don't have any public facilities, so they won't be included in this new dataset. Since I am only using one metric to determine the best neighborhood, I can just exclude the neighborhoods with 0 public facilities.

```python
neighborhoods = pf['neighborhood'].value_counts()
print('Number of public facilities by neighborhood')
print(neighborhoods)
```

```python
zeros = pd.Series(['Arlington','Arlington Heights','Chartiers','East Carnegie','Esplen','Friendship','Mount Oliver','New Homestead','North Point Breeze','North Shore','Northview Heights','Ridgemont','Saint Clair','Southshore','Spring Garden','Summer Hill'])
print("")
print('Neighborhoods with no public facilities:')
print(zeros)
```

# Graph of Public Facilities by neighborhood
I then graphed the 75 neighborhoods based on the number of public facilities they have. The graph is hard to read, so I cut out the bottom 70 neighborhoods (less than 15 public facilities) below.

```python
neighborhoods.plot(kind = 'bar', figsize = (75,30))
```

```python
top5 = neighborhoods[neighborhoods >= 15]
top5.plot(kind = 'bar', title = "Neighborhoods with the most public facilities in Pittsburgh")
plot.xlabel('Neighborhood')
plot.ylabel('Number of Public Facilities')
```

# The Best Neighborhood
Highland Park has the highest number of public facilities, making it the best Neighborhood in Pittsburgh. Squirrel Hill South was just one facility behind, with Perry North, Brookline, and the Strip District having a strong 15-20 public facilities each. The large amount of public facilities in Highland Park and Squirrel Hill South is most likely due to Highland Park's park and Squirrel Hill South's Schenley and Frick parks. In these parks, there are a lot of playgrounds, athletic fields, and other recreational sites that make up a lot of the data in WPRDC's dataset.
